# Plot LCCA for Five Sea Ice Concentration Products (Single Year)

- blue: Bremen-ASI
- magenta: NSIDC-NT2
- green: ERA5
- orange: OSI-401
- black: OSI-408
- cyan: MERRA-2

```python
import numpy as np
import os
import glob
import datetime as dt
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
def ReadLCCA(filename):
    '''
    Read data and LCCA value form txt files
    '''
    file = open(filename,'r')
    date = []       # peak date
    val = []        # peak value
    try:
        for line in file.readlines():
    #            print(line)
            if line[0]!='\n':
                date.append(dt.datetime.strptime(line.split()[0],"%Y%m%d"))                
                val.append(float(line.split()[1]))
    finally:
        file.close()
    
    return date,val
```

```python
def switch_path(type):
    pub_path = "C:\\Users\\kathy\\research\\Sea-ice-concentration\\data\\LCCA\\"
    
    switcher = {
        ASI: pub_path+"ASI_2002_2019\\",
        NT2: pub_path+"NT2_2002_2019\\",
        ERA5: pub_path+'ERA5_2002_2019\\',
        OSI401: pub_path+'OSI-401\\',
        OSI408: pub_path+'OSI-408\\',
        MERRA2: pub_path+'MERRA2_2002_2019\\'
    }
    
    return switcher.get(type, "nothing") 
```

```python
ASI = 'ASI-SIC-LCCA-'
NT2 = 'NT2-SIC-LCCA-'
ERA5 = 'ERA5-SIC-LCCA-'
OSI401 = 'OSI-SSMI-SIC-LCCA-'
OSI408 = 'OSI-AMSR2-SIC-LCCA-'
MERRA2 = 'MERRA2-SIC-LCCA-'
```

```python
year = '2002'
# year = '2003'
# year = '2004'
# year = '2005'
# year = '2006'
# year = '2007'
# year = '2008'
# year = '2009'
# year = '2010'

# year = '2011'
# year = '2012'
# year = '2013'
# year = '2014'
# year = '2015'
# year = '2016'
# year = '2017'
# year = '2018'
# year = '2019'



fname0 = switch_path(ASI) + ASI + year + '-KD.txt'
fname1 = switch_path(NT2) + NT2 + year + '-KD.txt'
fname2 = switch_path(ERA5) + ERA5 + year + '-KD.txt'
# fname3 = switch_path(OSI401) + OSI401 + year + '-KD.txt'
# fname4 = switch_path(OSI408) + OSI408 + year + '-KD.txt'
fname5 = switch_path(MERRA2) + MERRA2 + year + '-KD.txt'

```

```python
date0, LCCA0 = ReadLCCA(fname0)
date1, LCCA1 = ReadLCCA(fname1)
date2, LCCA2 = ReadLCCA(fname2)
# date3, LCCA3 = ReadLCCA(fname3)
# date4, LCCA4 = ReadLCCA(fname4)
date5, LCCA5 = ReadLCCA(fname5)
```

```python
from matplotlib.font_manager import FontProperties  
font1 = FontProperties(family='times new roman', size=26)
font2 = FontProperties(family='times new roman', size=22)
```

```python
fig,ax = plt.subplots(figsize=(14, 8))   
# ax = plt.gca()   
'''
blue: Bremen-ASI
magenta: NSIDC-NT2
green: ERA 5
orange: OSI-SSMI
black: OSI-AMSR2
cyan: MERRA-2
'''
b_line = plt.plot(date0, LCCA0,color = 'blue',linewidth=2.0,label = 'Bremen')
r_line = plt.plot(date1, LCCA1,color = 'magenta',linewidth=2.0,label = 'NSIDC')
g_line = plt.plot(date2, LCCA2,color = 'seagreen',linewidth=2.0,label = 'ERA5')
# o_line = plt.plot(date3, LCCA3,color = 'orange',linewidth=2.0,label = 'OSI-SSM/I')
# k_line = plt.plot(date4, LCCA4,color = 'black',linewidth=2.0,label = 'OSI-AMSR2')
c_line = plt.plot(date5, LCCA5,color = 'cyan',linewidth=2.0,label = 'MERRA-2')



plt.xlabel("Time",fontproperties = font1)
plt.ylabel("LICI (%)",fontproperties = font1)

x_begin = dt.datetime.strptime(str(year)+'06','%Y%m')
x_end = dt.datetime.strptime(str(year)+'11','%Y%m')

xtick = plt.xlim(x_begin,x_end)
plt.xticks(fontproperties = font2,rotation=0)

ytick = plt.yticks(np.arange(0,72,10),fontproperties = font2)  


## legend
plt.legend(loc=2,bbox_to_anchor=(0.02,0.98),numpoints=1)  # 添加图例
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext,fontproperties=font2)

# plt.savefig(year+'_six.png')
# plt.savefig(year+'_five.png')
plt.savefig(year+'_four.png')

```
